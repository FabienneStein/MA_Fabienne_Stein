```python
import numpy as np
import pandas as pd
# again bc of data privacy reasons I was not allowed to publish the dataset
df_or = pd.read_csv("time_series.csv")
df_train = pd.read_csv("train_df.csv")
df_test = pd.read_csv("test_df.csv")
# predict quarterly
time_horizon = 360
```

```python
en = df_or["Aktueller_Strombezug_kW"].to_numpy()
date = df_or["Datetime"].to_numpy()
date = np.array(date, dtype='datetime64')
print(en.mean())
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20, 6))
plt.plot(date, en, label='Energy Demand')
plt.xlabel('Date')
plt.ylabel('Energy Demand in kW')
plt.title("Energy Demand")

tick_positions = ['2016','2017', '2018', '2019', '2020', 
                    '2021', '2022', '2023']
plt.xticks(tick_positions, tick_positions)
plt.grid()
plt.legend()
plt.savefig("energy demand")
plt.show()
```
